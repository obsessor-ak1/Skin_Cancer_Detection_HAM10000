## HAM10000 Dataset
This dataset is a collection of skin lesion images collected across 10 classes consisting both benign and malignant (cancerous) lesion types.

```python
from exp_tools.data_utils import HAM10000Dataset, Explorer
```

### Loading the dataset

The following code loads the dataset as `torch.data.utils.Dataset` object. So that it also becomes helpful at the time of model training

```python
data = HAM10000Dataset(dataset_dir="../data/HMNIST")
len(data)
```

```python
data.label_map
```

```python
explorer = Explorer(data)
```

#### Examining the image distribution
The following code shows distribution of images over all labels:

```python
explorer.show_distribution()
```

It seems like the number of `nv` or Melanocytic nevi images are much larger than the others (more than 6000) and the `df` or  Dermatofibroma images are far less (around 100) images. Which shows there is a huge class imbalance in the dataset.

```python
explorer.show_distribution(criteria="dx_type")
```

For most of the images the labelling method used histopathology or follow up which are highly reliable methods.

## HAM10000 Dataset
This dataset is a collection of skin lesion images collected across 10 classes consisting both benign and malignant (cancerous) lesion types.

```python
from exp_tools.data_utils import HAM10000Dataset, Explorer
```

### Loading the dataset

The following code loads the dataset as `torch.data.utils.Dataset` object. So that it also becomes helpful at the time of model training

```python
data = HAM10000Dataset(dataset_dir="./data/HMNIST")
len(data)
```

The dataset has 10015 lesion images across following classes:

| Label     | Full Name                                          | Cancer?                      |
|-----------|----------------------------------------------------|------------------------------|
| **akiec** | Actinic keratoses and intraepithelial carcinoma    | ‚úÖ Pre-cancerous / malignant |
| **bcc**   | Basal cell carcinoma                               | ‚úÖ Malignant                 |
| **bkl**   | Benign keratosis-like lesions                      | ‚ùå Benign                    |
| **df**    | Dermatofibroma                                     | ‚ùå Benign                    |
| **nv**    | Melanocytic nevi                                   | ‚ùå Benign                    |
| **mel**   | Melanoma                                           | ‚úÖ Malignant                 |
| **vasc**  | Vascular lesions                                   | ‚ùå Benign                    |

The dataset is labelled using either of the following five strategies:

| dx_type    | Method Description                                                                 | Confidence Level |
|------------|--------------------------------------------------------------------------------------|------------------|
| **histo**     | Diagnosis confirmed by histopathology, i.e., tissue biopsy examined under a microscope | üî¨ Very High       |
| **confocal**  | Diagnosis made using reflectance confocal microscopy, a non-invasive optical imaging technique | üü° Medium          |
| **follow_up** | Diagnosis based on consistent clinical appearance over time without significant change | üü¢ High            |
| **consensus** | Diagnosis agreed upon by multiple expert dermatologists through visual inspection | üü° Medium          |
| **virtual**   | Diagnosis assigned through crowdsourced or web-based consensus, often without in-person evaluation | ‚ö™ Low             |


### Exploring the dataset
The following code explores the dataset and look at some of the images of specific type.

```python
explorer = Explorer(data)
```

#### Examining the image distribution
The following code shows distribution of images over all labels:

```python
explorer.show_distribution()
```

It seems like the number of `nv` or Melanocytic nevi images are much larger than the others (more than 6000) and the `df` or  Dermatofibroma images are far less (around 100) images. Which shows there is a huge class imbalance in the dataset.

```python
explorer.show_distribution(criteria="dx_type")
```

For most of the images the labelling method used histopathology or follow up which are highly reliable methods.
